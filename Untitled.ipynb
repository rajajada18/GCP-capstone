```python
from google.cloud import storage

project_id = "data-lake-412809"

# Create a storage client
client = storage.Client(project=project_id)

# Specify the bucket name
bucket_name = "myfirst-datalake"

# Get the bucket
bucket = client.get_bucket(bucket_name)

# Specify the blob (object) name within the bucket
blob_name = "StudentDetails.txt"

# Create a Blob object
blob = bucket.blob(blob_name)

# Now, you can perform various operations on the blob, such as uploading, downloading, deleting, etc.
# For example, uploading a file to the blob:
local_file_path = "StudentDetails.txt"


blob.download_to_filename(local_file_path )

```
